```python
import numpy as np
import pandas as pd
import edhec_risk_kit as erk
```

```python
ind_return = erk.get_ind_return()
ind_nfirms = erk.get_ind_nfirms()
ind_size = erk.get_ind_size()
```

```python
print(ind_return.shape)
print(ind_nfirms.shape)
print(ind_size.shape)
```

```python
ind_market_cap = ind_nfirms * ind_size
ind_market_cap.shape
```

```python
total_market_cap = ind_market_cap.sum(axis=1)
```

```python
total_market_cap.plot()
```

```python
ind_cap_weight = ind_market_cap.divide(total_market_cap, axis=0)
ind_cap_weight
```

```python
ind_cap_weight["1926"].sum(axis=1)
```

```python
ind_cap_weight[["Fin","Steel"]].plot()
```

```python
total_market_return = (ind_cap_weight * ind_return).sum(axis=1)
```

```python
total_market_return.plot()
```

```python
total_market_index = erk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize=(15,10))
```

```python
total_market_index["1980":].plot()
total_market_index["1980":].rolling(window=36).mean().plot()
```
